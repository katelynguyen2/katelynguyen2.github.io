```python
import pandas as pd 
import numpy as np
import altair as alt

data_url = "https://raw.githubusercontent.com/UIUC-iSchool-DataViz/is445_data/main/bfro_reports_fall2022.csv"
big = pd.read_csv(data_url)

big.head()
big.info()

big["date"] = pd.to_datetime(big["date"], errors="coerce")
big["Year"] = big["date"].dt.year
big["Season"] = big["season"].astype(str)

state_counts = big.groupby("state").size().reset_index(name="Count")
state_counts["state"] = state_counts["state"].astype(str)

brush = alt.selection_interval(encodings=['x','y'])

chart = alt.Chart(state_counts).mark_bar().encode(
    alt.X("state:N", sort="-y", title="State"),
    alt.Y("Count:Q", title="Number of Reports"),
    alt.Color("Count:Q", scale=alt.Scale(scheme="blues"), title="Report Count"),
    tooltip=[alt.Tooltip("state:N", title="State"),
             alt.Tooltip("Count:Q", title="Count")]
).properties(
    title="Bigfoot (BFRO) Reports by U.S. State",
    width=600,
    height=400
)

chart
```
